# PhraseMatcher

We use spaCy's PhraseMatcher to identify rows from our dataframe which contain euphemisms from our pre-defined list.

```python
import spacy.cli
spacy.cli.download('zh_core_web_md') # zh_core_web_md is the Chinese language download
```

```python
# prepare the dataframe for the search
df.rename(columns={'text': 'orig_text'}, inplace=True)

# Initialize a "keywords" column with an empty string for all rows
df['PET'] = ""

#df2.reset_index(drop=True, inplace=True) # This is for resetting the indices, which we may not want to do?
pd.reset_option('display.max_colwidth') # Remove text wrap
df
```

```python
import spacy
from spacy.matcher import PhraseMatcher
from tqdm import tqdm # module for displaying a progress bar

nlp = spacy.load("zh_core_web_md")
matcher = PhraseMatcher(nlp.vocab,attr="LOWER")

terms = euph_list # here, specify the list of phrases that PhraseMatcher should search for

patterns = [nlp.make_doc(text) for text in terms]
print(patterns)
matcher.add("TerminologyList", patterns)

#For each row, run matcher() on it and add to keywords column if match found; otherwise, drop() the row
with tqdm(total=num_rows) as pbar: # this loop is for the progress bar
    for index, row in df.iterrows():
        doc = nlp(row['orig_text'])
        matches = matcher(doc)
        if (len(matches) == 0):
            df = df.drop(labels=index, axis=0)
        else:
            for match_id, start, end in matches:
                span = doc[start:end]
                df.at[index, 'PET'] = span.text
        pbar.update(1)
        
#df.to_csv('output.csv')
```

### For forming the annotators' sample:

```python
df = pd.read_csv('random_always_euphs_12-1.csv', index_col = 0, encoding= 'utf-8')
types = df.type.unique()

# the DataFrame that will store the annotation samples
annotation_df = df.loc[df['type'] == -1]

# first add the 200 always_euph samples
index = 0
while (len(annotation_df) < 200):
    if (index >= len(types)): # if we reached the last type, restart
        index = 0
        continue
    currType = types[index]
    samples = df.loc[df['type'] == currType]
    if (len(samples) == 0): # if none left of this type, move on
        index += 1
        continue
    sample = samples.sample(n=1)
    annotation_df = annotation_df.append(sample)
    df = df.drop(sample.index[0])
    index += 1

print(len(annotation_df.type.unique()))
```

```python
df = pd.read_csv('random_sometimes_euphs_12-1.csv', index_col = 0, encoding= 'utf-8')
types = df.type.unique()

# now let's add the 400 sometimes_euph samples
index = 0
alternator = True # this will help get alternating samples of 0 and 1
while (len(annotation_df) < 600):
    if (index >= len(types)): # if we reached the last type, restart
        index = 0
        continue
    currType = types[index]
    samples = df.loc[df['type'] == currType]
    samples = samples.loc[df['is_euph'] == int(alternator)]
    if (len(samples) == 0): # if none left of this type, move on
        index += 1
        continue
    sample = samples.sample(n=1)
    annotation_df = annotation_df.append(sample)
    alternator = not alternator
    df = df.drop(sample.index[0])
    index += 1

print(len(annotation_df.type.unique()))
```
